```python
# 信号的基本运算
包括幅度变化、时移、翻转、累积、差分（或者近似微分）、求和、能量、功率、相加、相乘
```

```python
# 导入 需要的 library 库  
import numpy as np # 科学计算
import matplotlib.pyplot as plt # 画图
import scipy.signal as sg # 导入 scipy 的 signal 库 命名为 sg

import warnings
warnings.filterwarnings("ignore") # 去掉常规警告
```

## 幅度变化

```python
x = np.array([1,2,5,4,6,9]) # 基本信号
n = np.arange(len(x)) # 离散时间点序列
A = 2 # 幅度系数
fig, axs = plt.subplots(2, 1, figsize=(10, 5)) # 通过figsize调整图大小
plt.subplots_adjust(wspace = 0, hspace = 0.4) # 通过hspace调整子图间距
plt.subplot(211) # 绘制x(n)信号的子图
plt.stem(n, x, use_line_collection=True) # 绘制基本信号
plt.grid() # 显示网格
_ = plt.title('x[n]') # x[n]信号title

plt.subplot(212) # 绘制A*x(n)信号的子图
plt.stem(n, A*x, use_line_collection=True) # 绘制幅度变化的信号
plt.grid() # 显示网格
_ = plt.title(f'{A}*x[n]') # A*x[n]信号title
```

## 时移

```python
x = np.array([1,2,5,4,6,9]) # 基本信号
n = np.arange(len(x)) # 离散时间点序列
k = 4 # 时移量
fig, axs = plt.subplots(2, 1, figsize=(10, 5)) # 通过figsize调整图大小
plt.subplots_adjust(wspace = 0, hspace = 0.4) # 通过hspace调整子图间距
plt.subplot(211) # 绘制x(n)信号的子图
plt.stem(n, x, use_line_collection=True) # 绘制基本信号
plt.grid() # 显示网格
_ = plt.xticks(np.arange(0, len(x)+k+1, step=1.0))  # 横坐标间隔
_ = plt.title('x[n]') # x[n]信号title

plt.subplot(212) # 绘制A*x(n)信号的子图
plt.stem(np.append(np.zeros(k),n+4), np.append(np.zeros(k),x), use_line_collection=True) # 绘制时移变化的信号
plt.grid() # 显示网格
_ = plt.xticks(np.arange(0, len(x)+k+1, step=1.0))  # 横坐标间隔
_ = plt.title(f'x[n-{k}]') # x[n-k]信号title
```

## 翻转

```python
x = np.array([1,2,5,4,6,9]) # 基本信号
n = np.arange(len(x)) # 离散时间点序列
fig, axs = plt.subplots(2, 1, figsize=(10, 5)) # 通过figsize调整图大小
plt.subplots_adjust(wspace = 0, hspace = 0.4) # 通过hspace调整子图间距
plt.subplot(211) # 绘制x(n)信号的子图
plt.stem(n, x, use_line_collection=True) # 绘制基本信号
plt.grid() # 显示网格
_ = plt.xticks(np.arange(-len(x), len(x)+1, step=1.0))  # 横坐标间隔
_ = plt.title('x[n]') # x[n]信号title

plt.subplot(212) # 绘制A*x(n)信号的子图
plt.stem(-n[::-1], x[::-1], use_line_collection=True) # 绘制翻转变化的信号
plt.grid() # 显示网格
_ = plt.xticks(np.arange(-len(x), len(x)+1, step=1.0))  # 横坐标间隔
_ = plt.title('x[-n]') # x[-n]信号title
```

## 累积

```python
x = np.array([1,2,5,4,6,9]) # 基本信号
n = np.arange(len(x)) # 离散时间点序列
print(f'信号累积和为：{np.sum(x)}')
```

## 差分（或者近似微分）

```python
x = np.array([1,2,5,4,6,9]) # 基本信号
n = np.arange(len(x)) # 离散时间点序列
print(f'信号差分结果为：{np.diff(x)}')
```

## 求和

```python
x = np.array([1,2,5,4,6,9]) # 基本信号
n = np.arange(len(x)) # 离散时间点序列
n1 = 1
n2 = 4
print(f'信号求和n[{n1}]~n[{n2}]为：{np.sum(x[n1:n2+1])}')
```

## 能量

```python
x = np.array([1,2,5,4,6,9]) # 基本信号
n = np.arange(len(x)) # 离散时间点序列
print(f'信号能量为：{np.sum(np.abs(x)**2)}')
```

## 功率

```python
x = np.array([1,2,5,4,6,9]) # 基本信号
n = np.arange(len(x)) # 离散时间点序列
N = len(x)
print(f'信号功能为：{np.sum(np.abs(x)**2)/N}')
```

## 相加

```python
x1 = np.array([1,2,5,4,6,9]) # 基本信号
x2 = np.array([1,4,9,2,7,9]) # 基本信号
n = np.arange(len(x1)) # 离散时间点序列
fig, axs = plt.subplots(3, 1, figsize=(10, 10)) # 通过figsize调整图大小
plt.subplots_adjust(wspace = 0, hspace = 0.4) # 通过hspace调整子图间距
plt.subplot(311) # 绘制x(n)信号的子图
plt.stem(n, x1, use_line_collection=True) # 绘制基本信号x1
plt.grid() # 显示网格
_ = plt.xticks(np.arange(0, len(x1)+1, step=1.0))  # 横坐标间隔
_ = plt.title('x1[n]') # x1[n]信号title

plt.subplot(312) # 绘制A*x(n)信号的子图
plt.stem(n, x2, use_line_collection=True) # 绘制基本信号x2
plt.grid() # 显示网格
_ = plt.xticks(np.arange(0, len(x2)+1, step=1.0))  # 横坐标间隔
_ = plt.title('x2[n]') # x2[n]信号title

plt.subplot(313) # 绘制A*x(n)信号的子图
plt.stem(n, x1+x2, use_line_collection=True) # 绘制相加信号x1+x2
plt.grid() # 显示网格
_ = plt.xticks(np.arange(0, len(x1)+1, step=1.0))  # 横坐标间隔
_ = plt.title('x1[n]+x2[n]') # x1[n]+x2[n]信号title
```

## 相乘

```python
x1 = np.array([1,2,5,4,6,9]) # 基本信号
x2 = np.array([1,4,9,2,7,9]) # 基本信号
n = np.arange(len(x1)) # 离散时间点序列
fig, axs = plt.subplots(3, 1, figsize=(10, 10)) # 通过figsize调整图大小
plt.subplots_adjust(wspace = 0, hspace = 0.4) # 通过hspace调整子图间距
plt.subplot(311) # 绘制x(n)信号的子图
plt.stem(n, x1, use_line_collection=True) # 绘制基本信号x1
plt.grid() # 显示网格
_ = plt.xticks(np.arange(0, len(x1)+1, step=1.0))  # 横坐标间隔
_ = plt.title('x1[n]') # x1[n]信号title

plt.subplot(312) # 绘制A*x(n)信号的子图
plt.stem(n, x2, use_line_collection=True) # 绘制基本信号x2
plt.grid() # 显示网格
_ = plt.xticks(np.arange(0, len(x2)+1, step=1.0))  # 横坐标间隔
_ = plt.title('x2[n]') # x2[n]信号titl

plt.subplot(313) # 绘制A*x(n)信号的子图
plt.stem(n, x1*x2, use_line_collection=True) # 绘制相乘信号x1*x2
plt.grid() # 显示网格
_ = plt.xticks(np.arange(0, len(x1)+1, step=1.0))  # 横坐标间隔
_ = plt.title('x1[n]*x2[n]') # x1[n]*x2[n]信号title
```
